## Data Transformation and Splitting Concepts

**Data Transformation:**

Data transformation is the process of converting data from one format or structure into another. In machine learning, this is often done to:

*   **Improve model performance:** Some algorithms perform better with transformed data (e.g., scaling numerical features).
*   **Meet algorithm requirements:** Certain algorithms require specific data formats (e.g., categorical features need to be encoded).
*   **Reduce dimensionality:** Techniques like PCA can transform data into a lower-dimensional space while retaining important information.
*   **Handle skewed data:** Transformations like logarithmic or square root can make skewed data more symmetrical.

Common data transformation techniques include:

*   **Scaling:** Normalizing or standardizing numerical features (e.g., Min-Max scaling, StandardScaler).
*   **Encoding:** Converting categorical features into numerical representations (e.g., One-Hot Encoding, Label Encoding).
*   **Feature Engineering:** Creating new features from existing ones.
*   **Handling outliers:** Transforming or removing extreme values.

**Data Splitting:**

Data splitting is the process of dividing a dataset into multiple subsets, typically for training and evaluating a machine learning model. The most common split is into:

*   **Training set:** Used to train the model.
*   **Testing set:** Used to evaluate the model's performance on unseen data.

Sometimes, a third set is also used:

*   **Validation set:** Used to tune the model's hyperparameters and prevent overfitting on the training set.

The purpose of data splitting is to:

*   **Evaluate model generalization:** Assess how well the model performs on data it hasn't seen during training.
*   **Prevent overfitting:** Ensure the model doesn't simply memorize the training data but learns general patterns.

Common data splitting techniques include:

*   **Hold-out method:** A simple split into training and testing sets.
*   **Cross-validation:** Dividing the data into multiple folds and training/testing the model on different combinations of folds.

Splitting the data before any data transformation or preprocessing steps that use information from the entire dataset is crucial to prevent data leakage.

```python
# Identify outliers using box plots
pit.figure(figsize=(10, 6))
sms.boxplot(x=df['Age'])
pit.title('Box plot of Age')
pit.show()

pit.figure(figsize=(10, 6))
sms.boxplot(x=df['Fare'])
pit.title('Box plot of Fare')
pit.show()

# Remove outliers using IQR method
Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)
IQR = Q3 - Q1
df = df[~((df['Age'] < (Q1 - 1.5 * IQR)) | (df['Age'] > (Q3 + 1.5 * IQR)))]

Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)
IQR = Q3 - Q1
df = df[~((df['Fare'] < (Q1 - 1.5 * IQR)) | (df['Fare'] > (Q3 + 1.5 * IQR)))]

print("Outliers removed. Updated DataFrame info:")
df.info()
```
